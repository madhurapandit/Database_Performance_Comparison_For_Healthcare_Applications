```python
def movecol(df, cols_to_move=[], ref_col='', place='After'):
    
    cols = df.columns.tolist()
    if place == 'After':
        seg1 = cols[:list(cols).index(ref_col) + 1]
        seg2 = cols_to_move
    if place == 'Before':
        seg1 = cols[:list(cols).index(ref_col)]
        seg2 = cols_to_move + [ref_col]
    
    seg1 = [i for i in seg1 if i not in seg2]
    seg3 = [i for i in cols if i not in seg1 + seg2]
    
    return(df[seg1 + seg2 + seg3])


state_iss_age_tob_dep_num_yr = movecol(state_iss_age_tob_dep_num_yr, 
                                       cols_to_move=['plan_id','plan_name'], 
                                       ref_col='state_code',
                                       place='Before')

state_iss_age_tob_dep_num_yr = movecol(state_iss_age_tob_dep_num_yr, 
                                       cols_to_move=['business_year'], 
                                       ref_col='issuer_name',
                                       place='After')
state_iss_age_tob_dep_num_yr.head()
```

```python
start = '1940-01-01' 
end = '2021-12-31'

np.random.seed(4)
val1 = np.random.choice(['Male','Female'],nrows)
val2 = pd.to_datetime(np.random.randint(pd.Timestamp(start).value, pd.Timestamp(end).value,nrows, dtype=np.int64)).strftime('%Y/%m/%d')

## create age column
today = datetime.today()
ages = []
for value in val2:
    date = datetime.strptime(value, "%Y/%m/%d").date()
    age = today.year - date.year - ((today.month, today.day) < (date.month, date.day))
    ages.append(age)
    
df = pd.DataFrame({'Gender':val1, 'DOB':val2, 'Age':ages})
df.head()
```

```python
from faker import Faker

val4 = []
val5 = []
val6 = []
val7 = [] 

fake = Faker()
Faker.seed(4)

for i in range(nrows):
    text = fake.address() # address
    v4 = text.split("\n")[0] #street
    val4.append(v4)
    v5 = (text.split("\n")[1].split(", "))[0] # city
    val5.append(v5)
    v6 = ((text.split("\n")[1]).split(","))[-1].strip().split(" ")[0] # state
    val6.append(v6)
    v7 = ((text.split("\n")[1]).split(","))[-1].strip().split(" ")[1] # zipcode
    val7.append(v7)
```

```python
df['Street'] = val4
df['City'] = val5
df['State'] = val6
df['Zip'] = val7

df.head()
```

```python
import random
random.seed(9)
def random_phone_num_generator():
    first = str(random.randint(100, 999))
    second = str(random.randint(1, 888)).zfill(3)
    last = (str(random.randint(1, 9998)).zfill(4))
    while last in ['1111', '2222', '3333', '4444', '5555', '6666', '7777', '8888']:
        last = (str(random.randint(1, 9998)).zfill(4))
    return '{}-{}-{}'.format(first, second, last)

home_num = []
for i in range(0, nrows):
    home_num.append(random_phone_num_generator())
    
df['Home_Contact'] = home_num
df.head()
```

```python
work_num = []

random.seed(3)

for i in range(0, nrows):
    work_num.append(random_phone_num_generator())
    
df['Work_Contact'] = work_num
df.head()
```

```python
import random
import string

random.seed(6)

def random_char(y):
       return ''.join(random.choice(string.ascii_lowercase) for x in range(y))
    
email = []
for i in range(nrows):
    e = random_char(7)+"@xyz.com"
    email.append(e)
    
df['EmailID'] = email
df.head()
```

```python
def movecol(df, cols_to_move=[], ref_col='', place='After'):
    
    cols = df.columns.tolist()
    if place == 'After':
        seg1 = cols[:list(cols).index(ref_col) + 1]
        seg2 = cols_to_move
    if place == 'Before':
        seg1 = cols[:list(cols).index(ref_col)]
        seg2 = cols_to_move + [ref_col]
    
    seg1 = [i for i in seg1 if i not in seg2]
    seg3 = [i for i in cols if i not in seg1 + seg2]
    
    return(df[seg1 + seg2 + seg3])


df = movecol(df, cols_to_move=['PatientID','First_Name', 'Last_Name'], 
             ref_col='Gender', place='Before')

df = movecol(df, cols_to_move=['Date_of_visit', 'No_of_Dependents', 'Tobacco_preference'], 
             ref_col='EmailID', place='After')

df.head()
```
